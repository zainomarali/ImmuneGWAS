```python
out_files =[i.split("/")[-1] for i in files]

out_files = ["sorted_" + i for i in out_files]

out_files = [paths['eqtl_tokyo'] + i for i in out_files]
```

```python
# need to make the chromosome columns all integers and sort by variant position 

for i, n in enumerate(files):
    x = pd.read_csv(n, sep='\t')
    x['int_chrom'] = x.CHR.apply(lambda x:int(x.split("r")[-1]))
    x = x.sort_values(by=['int_chrom', 'Variant_position_start'])
    name = out_files[i]
    x.to_csv(name, sep='\t', index=False)
```
